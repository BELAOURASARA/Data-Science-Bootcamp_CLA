```python
import numpy as np
from PIL import Image

#reading the image and transposing it
image=Image.open('flower.png')
image.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
#extracting a square matrix from the multidimensional array
mat = np.array(image) #transforme the image to a numpy array
mat=mat.tolist() #transforme the array to a list so we can delete some elements easily
del mat[0][3:-1]
mat=mat[0]
mat = np.array(mat)
```

```python
#transposing the image
mat.T
```

```python
#check if the matrix is diagonal
rows=len(mat[0])
colomns=len(mat)
if rows==colomns : #check first if it is a square matrix
    
    n=np.count_nonzero(mat - np.diag(np.diagonal(mat)))
    if n==0 : 
        print("the matrix is diagonal")
    else : 
        print("the matrix is not diagonal")
    #or we can use directly : 
    if (np.allclose(mat, np.diag(np.diag(mat)))):
        print("the matrix is diagonal")
    else : 
        print("the matrix is not diagonal") 
        
else: 
    print("the matrix is not a square matrix , so we can't check if it is diagonal")

```

```python
#check if the matrix is triagonal
rows=len(mat[0])
colomns=len(mat)
if rows==colomns : #check first if it is a square matrix
    
    if (np.allclose(mat, np.tril(mat))):
        print("the matrix is lower triagonal")
    else: 
        if  (np.allclose(mat, np.triu(mat))) : 
            print("the matrix is upper triagonal")
        else:   
            print("the matrix is not triagonal")
        
else: 
    print("the matrix is not a square matrix , so we can't check if it is triagonal")
```

```python
#check if the matrix is symmetric
if (mat==mat.T).all(): 
    print("the matrix is symmetric")
else: 
    print("the matrix is not symmetric")
```
